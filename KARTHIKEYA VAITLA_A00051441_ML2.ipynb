# Image Classification Project: Imagenette Dataset Analysis

## 1. Dataset Selection
Dataset: Imagenette (Public Image Classification Dataset)
- Contains images from 10 ImageNet classes
- Designed for quick experiments and transfer learning
- Provides a diverse set of natural images

## 2. Data Import and Preprocessing

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
import tensorflow_datasets as tfds
import pandas as pd

# Step 2: Loading and Preprocessing the Flowers Dataset
dataset_name = "imagenette"
(train_ds, val_ds), ds_info = tfds.load(
    dataset_name,
    split=['train[:80%]', 'train[80%:]'],
    as_supervised=True,
    with_info=True
)

num_classes = ds_info.features['label'].num_classes
class_names = ds_info.features['label'].names
print(f"Number of classes: {num_classes}")
print(f"Class names: {class_names}")

# Define preprocessing parameters
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

## 3. Neural Network Design
### 3.1 Simple Convolutional Neural Network (CNN)

# Create preprocessing function for training (with augmentation)
def preprocess_train(image, label):
    if len(image.shape) == 2:  # Grayscale image (2D)
        image = tf.expand_dims(image, axis=-1)  # Add a channel dimension
    elif len(image.shape) == 3 and image.shape[-1] == 1:  # Single-channel image
        image = tf.image.grayscale_to_rgb(image)  # Convert to 3 channels
    # Resize the image
    image = tf.image.resize(image, IMG_SIZE)
    # Data augmentation
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_brightness(image, 0.1)
    # Convert to float and normalize to [0,1]
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# Preprocessing for validation (no augmentation)
def preprocess_val(image, label):
    if len(image.shape) == 2:  # Grayscale image (2D)
        image = tf.expand_dims(image, axis=-1)  # Add a channel dimension
    elif len(image.shape) == 3 and image.shape[-1] == 1:  # Single-channel image
        image = tf.image.grayscale_to_rgb(image)  # Convert to 3 channels
    # Resize the image
    image = tf.image.resize(image, IMG_SIZE)
    # Convert to float and normalize to [0,1]
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# Apply preprocessing to datasets
train_ds = train_ds.map(preprocess_train, num_parallel_calls=AUTOTUNE)
train_ds = train_ds.shuffle(1000).batch(BATCH_SIZE).prefetch(AUTOTUNE)

val_ds = val_ds.map(preprocess_val, num_parallel_calls=AUTOTUNE)
val_ds = val_ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)

# Save preprocessed datasets to disk
train_ds.save('train_ds')
val_ds.save('val_ds')
print("Preprocessed datasets saved to disk.")

# Load preprocessed datasets from disk
train_ds = tf.data.Dataset.load('train_ds')
val_ds = tf.data.Dataset.load('val_ds')
print("Preprocessed datasets loaded from disk.")

# Confirm data shapes and ranges
print("Checking data:")
for images, labels in train_ds.take(1):
    print(f"Image batch shape: {images.shape}")
    print(f"Label batch shape: {labels.shape}")
    print(f"Image min-max range: {tf.reduce_min(images).numpy():.4f}-{tf.reduce_max(images).numpy():.4f}")
    print(f"Labels range: {tf.reduce_min(labels).numpy()}-{tf.reduce_max(labels).numpy()}")
    labels_flat = labels.numpy().flatten()
    print(f"Label distribution: {np.bincount(labels_flat)}")

# Display sample images to verify data
plt.figure(figsize=(12, 8))
for images, labels in train_ds.take(1):
    for i in range(min(9, len(images))):
        ax = plt.subplot(3, 3, i + 1)
        image = np.clip(images[i].numpy(), 0.0, 1.0)
        plt.imshow(image)
        plt.title(f"Class: {class_names[labels[i].numpy()]}")
        plt.axis("off")
plt.tight_layout()
plt.show()

# Build a simple CNN model
print("Building CNN model...")
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])

# Model Layers Explanation:
# 1. Convolutional Layers: Extract image features
# 2. Max Pooling: Reduce spatial dimensions
# 3. Flatten: Convert 2D feature maps to 1D
# 4. Dense Layers: Classification



# Compile the model
model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
print("Model summary:")
model.summary()

# Train the model
print("Training CNN model...")
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    verbose=1
)

# Evaluate the model
print("Evaluating CNN model...")
test_loss, test_acc = model.evaluate(val_ds, verbose=1)
print(f"\n✅ CNN Model Accuracy: {test_acc:.2%}")

# Load MobileNetV2 as the base model
print("Building MobileNetV2 model...")
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)

# Freeze the base model
base_model.trainable = False

# Build the transfer learning model
inputs = tf.keras.Input(shape=(224, 224, 3))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)
model = tf.keras.Model(inputs, outputs)

# Compile the model
model.compile(
    optimizer=optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
print("Model summary:")
model.summary()

# Train the model
print("Training MobileNetV2 model...")
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    verbose=1
)
## 4. Model Training and Evaluation
# Evaluate the model
print("Evaluating MobileNetV2 model...")
test_loss, test_acc = model.evaluate(val_ds, verbose=1)
print(f"\n✅ MobileNetV2 Model Accuracy: {test_acc:.2%}")

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Display predictions with confidence scores
images_batch, labels_batch = next(iter(val_ds))
predictions = model.predict(images_batch)
predicted_classes = np.argmax(predictions, axis=1)

plt.figure(figsize=(12, 12))
for i in range(min(9, len(images_batch))):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images_batch[i])

    true_label = class_names[labels_batch[i]]
    pred_label = class_names[predicted_classes[i]]
    confidence = predictions[i][predicted_classes[i]] * 100

    color = 'green' if true_label == pred_label else 'red'
    plt.title(f"True: {true_label}\nPred: {pred_label}\nConf: {confidence:.1f}%",
              color=color, fontsize=9)
    plt.axis("off")

plt.tight_layout()
plt.show()

def predict_single_image(img_path):
    # Load and preprocess the image
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=IMG_SIZE)
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) / 255.0  # Normalize to [0, 1]

    # Make prediction
    prediction = model.predict(img_array)
    predicted_class = np.argmax(prediction, axis=1)[0]
    confidence = prediction[0][predicted_class] * 100

## 5. Training Process Visualization
    # Display result
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.title(f"Predicted: {class_names[predicted_class]}\nConfidence: {confidence:.1f}%")
    plt.axis('off')
    plt.show()

    # Print top 3 predictions
    top_3_idx = np.argsort(prediction[0])[-3:][::-1]
    print("Top 3 predictions:")
    for idx in top_3_idx:
        print(f"{class_names[idx]}: {prediction[0][idx]*100:.1f}%")

    return class_names[predicted_class], confidence

# Step 8: Computing Accuracy on Test Data and Saving the Model

# Evaluate the model on test data
print("Computing accuracy on test data...")
test_loss, test_acc = model.evaluate(val_ds, verbose=1)
print(f"\n✅ Test Accuracy: {test_acc:.2%}")

# Save the model
model.save('imagenette_classifier_mobilenetv2.h5')
print("Model saved as 'imagenette_classifier_mobilenetv2.h5'")

```

### Python Code Questions
Q1: Why use `sparse_categorical_crossentropy` as the loss function?
A1: This loss function is ideal for integer-encoded labels (like our dataset) where each sample belongs to one class. It's computationally efficient and works well with multi-class classification.

Q2: What does `GlobalAveragePooling2D()` do in the transfer learning model?
A2: It reduces the spatial dimensions of feature maps by taking the average of each feature map, effectively creating a fixed-size representation regardless of input image size.

### Machine Learning Concept Questions
Q1: What is transfer learning, and why is it beneficial?
A1: Transfer learning allows using a pre-trained model (like MobileNetV2) trained on a large dataset to solve a different but related problem. It helps overcome limited training data and reduces training time.

Q2: How do convolutional layers differ from dense layers in neural networks?
A2: Convolutional layers are specialized for processing grid-like data (images) by applying filters to detect local features, while dense layers connect every neuron and are better for global pattern recognition.
